# Implicit : Lobatto : ADAM : Hubber Loss : Numeric Gradient : Setting_2 , With_Energy_error and Without_Energy_Error

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_function(ax, y, start_pt=0, title=""):
    x = range(start_pt, start_pt + len(y))
    ax.plot(x, y, label="Error (JAX)")
    ax.legend()
    ax.grid()
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Error')
    ax.set_yscale('log')  # Set the y-axis to logarithmic scale
    ax.set_title(title)
    for tick in ax.get_xticklabels():
        tick.set_rotation(45)
```

# With And Without Together 

```python
def create_plots(first_row_prefix, second_row_prefix, error_path, energy_path, validation_path, 
                 perturbed_error_path=None, perturbed_energy_path=None, perturbed_validation_path=None):
    
    # Load Files for the first set (Original)
    S1_Total_Error = np.genfromtxt(error_path, delimiter=',', invalid_raise=False, filling_values=np.nan) 
    S1_Energy_Error = np.genfromtxt(energy_path, delimiter=',', invalid_raise=False, filling_values=np.nan) 
    S1_Validation_Error = np.genfromtxt(validation_path, delimiter=',', invalid_raise=False, filling_values=np.nan) 

    # Subplots
    if perturbed_error_path and perturbed_energy_path and perturbed_validation_path:
        fig, axs = plt.subplots(2, 3, figsize=(16, 6))  # Two rows for original and second row with customizable prefix
    else:
        fig, axs = plt.subplots(1, 3, figsize=(16, 6))  # One row for original only
    axs = axs.flatten()

    ##### First Row (Customizable prefix for labels)
    
    s1_error = S1_Total_Error[0:]
    l = len(s1_error)
    plot_function(axs[0], s1_error, 0, f"{first_row_prefix} Total Error")

    s1_energy = S1_Energy_Error[0:l]
    plot_function(axs[1], s1_energy, 0, f"{first_row_prefix} Energy Error")
    
    s1_validation = S1_Validation_Error[0:l]
    plot_function(axs[2], s1_validation, 0, f"{first_row_prefix} Validation Error")

    ##### Second Row (Customizable prefix for labels)
    if perturbed_error_path and perturbed_energy_path and perturbed_validation_path:
        P_S1_Total_Error = np.genfromtxt(perturbed_error_path, delimiter=',', invalid_raise=False, filling_values=np.nan)
        P_S1_Energy_Error = np.genfromtxt(perturbed_energy_path, delimiter=',', invalid_raise=False, filling_values=np.nan) 
        P_S1_Validation_Error = np.genfromtxt(perturbed_validation_path, delimiter=',', invalid_raise=False, filling_values=np.nan) 
        
        p_s1_error = P_S1_Total_Error[0:l]
        plot_function(axs[3], p_s1_error, 0, f"{second_row_prefix} Total Error")

        p_s1_energy = P_S1_Energy_Error[0:l]
        plot_function(axs[4], p_s1_energy, 0, f"{second_row_prefix} Energy Error")
        
        p_s1_validation = P_S1_Validation_Error[0:l]
        plot_function(axs[5], p_s1_validation, 0, f"{second_row_prefix} Validation Error")
    

    # Set the main title for the plot
    fig.suptitle(f'Error Convergence plot: step = 100 : istep = 10', fontsize=10)

    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for the super title
    plt.show()
```

# Lobatto 3A 3B : Original + Perturbed 

```python

# Call the plotting function 
create_plots(
    "Original",  # First row prefix
    "Perturbed", # Second row prefix
    'With_Energy/One-batch-Lobatto-Outputs/Error.txt', 
    'With_Energy/One-batch-Lobatto-Outputs/Energy.txt',
    'With_Energy/One-batch-Lobatto-Outputs/Validation.txt',
    'With_Energy/One-batch-PerturbedLobatto-Outputs/Error.txt', 
    'With_Energy/One-batch-PerturbedLobatto-Outputs/Energy.txt',
    'With_Energy/One-batch-PerturbedLobatto-Outputs/Validation.txt'
)

create_plots(
    "Original",  
    "Perturbed", 
    'Without_Energy/One-batch-Lobatto-Outputs/Error.txt', 
    'Without_Energy/One-batch-Lobatto-Outputs/Energy.txt',
    'Without_Energy/One-batch-Lobatto-Outputs/Validation.txt',
    'Without_Energy/One-batch-PerturbedLobatto-Outputs/Error.txt', 
    'Without_Energy/One-batch-PerturbedLobatto-Outputs/Energy.txt',
    'Without_Energy/One-batch-PerturbedLobatto-Outputs/Validation.txt',
)
```
