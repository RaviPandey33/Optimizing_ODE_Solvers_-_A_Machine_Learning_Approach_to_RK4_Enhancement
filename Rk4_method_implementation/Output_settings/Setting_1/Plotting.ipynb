```python
import matplotlib.pyplot as plt

def plot_function(ax, y, start_pt=0, title=""):
    x = range(start_pt, start_pt + len(y))
    ax.plot(x, y, label="Error (JAX)", marker='o')
    ax.legend()
    ax.grid()
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Error')
    ax.set_title(title)
```

```python
# Not-Perturbed

import numpy as np
S1_output = np.loadtxt('S1_Error.txt')

#Subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 10))
axs = axs.flatten()

new_s1 = S1_output[0:3500]
plot_function(axs[0], new_s1, 0, "Epoch (0-3500)")

new_s1 = S1_output[3500:4000]
plot_function(axs[1], new_s1, 3500, "Epoch (3500-4000)")

new_s1 = S1_output[4000:10000]
plot_function(axs[2], new_s1, 4000, "Epoch (4000-10k)")

new_s1 = S1_output[10000:100000]
plot_function(axs[3], new_s1, 10000, "Epoch (10k-100k)")

new_s1 = S1_output[0:100000]
plot_function(axs[4], new_s1, 10000, "Complete Epochs")

plt.tight_layout()
plt.show()
```
