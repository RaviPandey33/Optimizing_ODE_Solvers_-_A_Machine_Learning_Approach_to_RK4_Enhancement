# Implicit : Lobatto : ADAM : Setting_1 , With_Energy_error and Without_Energy_Error

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_function(ax, y, start_pt=0, title=""):
    x = range(start_pt, start_pt + len(y))
    ax.plot(x, y, label="Error (JAX)", marker='o')
    ax.legend()
    ax.grid()
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Error')
    ax.set_title(title)
    for tick in ax.get_xticklabels():
        tick.set_rotation(45)
```

```python
print("With Energy Error")

# Original Weights and Perturbed Weights
# Load Files
S1_Total_Error_With_Energy = np.loadtxt('With_Energy/S1_Outputs/Error.txt')
S1_Energy_Error_With_Energy = np.loadtxt('With_Energy/S1_Outputs/Energy.txt')
S1_Validation_Error_With_Energy = np.loadtxt('With_Energy/S1_Outputs/Validation.txt')

P_S1_Total_Error_With_Energy = np.loadtxt('With_Energy/S1_Outputs/Perturbed_Error.txt')
P_S1_Energy_Error_With_Energy = np.loadtxt('With_Energy/S1_Outputs/Perturbed_Energy.txt')
P_S1_Validation_Error_With_Energy = np.loadtxt('With_Energy/S1_Outputs/Perturbed_Validation.txt')

#Subplots
fig, axs = plt.subplots(2, 3, figsize=(12, 6))
axs = axs.flatten()

##### Original
s1_error = S1_Total_Error_With_Energy[0:]
l = len(s1_error)
plot_function(axs[0], s1_error, 0, "Total Error")

s1_validation = S1_Validation_Error_With_Energy[0:l]
plot_function(axs[1], s1_validation, 0, "Validation Error")

s1_energy = S1_Energy_Error_With_Energy[0:l]
plot_function(axs[2], s1_energy, 0, "Energy Error")

##### Perturbed
p_s1_error = P_S1_Total_Error_With_Energy[0:]
l = len(p_s1_error)
print(l)
plot_function(axs[3], p_s1_error, 0, "Perturbed Total Error")

p_s1_validation = P_S1_Validation_Error_With_Energy[0:l]
plot_function(axs[4], p_s1_validation, 0, "Perturbed Validation Error")

p_s1_energy = P_S1_Energy_Error_With_Energy[0:l]
plot_function(axs[5], p_s1_energy, 0, "Perturbed Energy Error")

fig.suptitle('Comparison of Original and Perturbed Weights Errors Across Different Metrics. (Implicit, Lobatto 3AB : 4th Order, BFGS, Finite_Difference approx., Batched(number of batches = 1) )', fontsize=10)

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for the super title

# # Saving the plot
# plt.savefig('comparison_plot.png', dpi=300)

plt.show()
```
